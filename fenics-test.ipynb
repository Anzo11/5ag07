```
p0 = Point(-1.,-1.)
p1 = Point(1.,1.)
o = Point(0.,0.)
rect = Rectangle(p0,p1)
circle = Circle(o,.2)
geom = rect - circle
mesh = generate_mesh(geom,50)
plot(mesh)
```

```
plt.figure()
U = FunctionSpace(mesh,'P',1)
u = Function(U)
v = TestFunction(U)
ut = TrialFunction(U)
a = inner(grad(ut),grad(v))*dx
L = v*dx
A,b  = assemble_system(a,L)
solve(A,u.vector(), b)
plot(u)
```
